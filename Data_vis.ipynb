```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
#get data
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
#sort ordinal data
cut_kat = ["Fair","Good","Very Good", "Premium", "Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories=cut_kat, ordered = True))
df.head()
```

```python
sns.barplot(x="cut", y="price", hue ="color", data = df);
#visualize data
```

```python
sns.catplot(x="cut",y="price", data=df);
```

```python
df.groupby(["cut","color"])["price"].mean()
```

```python
sns.distplot(df.price,bins = 1000);
```

```python
sns.distplot(df.price,hist=False);
```

```python
sns.distplot(df.price,bins = 1000, kde=False);
```

```python
sns.kdeplot(df.price, shade = True);
```

```python
(sns
.FacetGrid(df,hue="cut", height=5, xlim = (0,15000))
.map(sns.kdeplot,"price",shade=True)
.add_legend());
```

```python
sns.catplot(x="cut",y="price",hue="color",kind="point",data=df);
#colors and cut affects price 
```
